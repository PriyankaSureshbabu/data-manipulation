```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np

data_blt = pd.read_csv("C:/Users/priya/OneDrive/Desktop/women+/data-manipulation/dataset/baltimore_9-24-2016_9-30-2017.csv")
data_bos = pd.read_csv("C:/Users/priya/OneDrive/Desktop/women+/data-manipulation/dataset/boston_9-24-2016_9-30-2017.csv")
data_nyc = pd.read_csv("C:/Users/priya/OneDrive/Desktop/women+/data-manipulation/dataset/new-york_9-24-2016_9-30-2017.csv")
data_phl = pd.read_csv("C:/Users/priya/OneDrive/Desktop/women+/data-manipulation/dataset/philadelphia_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
# Baltimore data cleanup
for col in data_blt.columns:
    pct_missing = np.mean(data_blt[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
data_blt.head()

#modifying Repack column values

data_blt["Repack"] = data_blt["Repack"].replace({"N":False , "E":True})

data_blt.head(3)

#shape

data_blt.shape

data_blt = data_blt.drop(['Type', 'Grade', 'Origin District', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode', 'Sub Variety', 'Color'], axis=1)



data_blt.shape

data_blt.head()
```

```python
# New York data cleanup

for col in data_nyc.columns:
    pct_missing = np.mean(data_nyc[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
data_nyc["Repack"] = data_nyc["Repack"].replace({"N":False , "E":True})

data_nyc.head(3)

data_nyc.shape

data_nyc = data_nyc.drop(['Type', 'Sub Variety', 'Grade', 'Origin District', 'Color', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode'], axis=1)

data_nyc.head()

data_nyc.shape
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
data_NorthEast = pd.concat([data_blt, data_bos, data_nyc, data_phl], keys=["Baltimore", "Boston", "New York", "Philadelphia"])

data_NorthEast
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
data_NorthEast1 = data_NorthEast.drop(['Repack'], axis=1)

data_NorthEast1['Unit of Sale'] = data_NorthEast1['Unit of Sale'].fillna('UNKNOWN')
data_NorthEast1

data_NorthEast1_grp = data_NorthEast1.groupby('Unit of Sale')

data_NorthEast1_grp

data_NorthEast1_grp.first()

agg_func_list = ['mean', 'min', 'max']

data_NorthEast1_grp.agg(agg_func_list) 
```

```python
data_NorthEast2 = data_NorthEast.drop(['Repack'], axis=1)

data_NorthEast2['Variety'] = data_NorthEast1['Variety'].fillna('UNKNOWN')
data_NorthEast2

data_NorthEast2_grp = data_NorthEast2.groupby('Variety')

data_NorthEast2_grp

data_NorthEast2_grp.first()

data_pumpkin = data_NorthEast2['Variety'].value_counts()
data_pumpkin

# df = data_pumpkin.rename_axis('Variety').reset_index(name='counts')
# print (df)

df = data_pumpkin.rename_axis('Variety').to_frame('counts')
print (df)

agg_func_list = ['mean']

df.agg(agg_func_list)
```
